### Zuordnung von Ländern zu Kontinenten

Mithilfe eines Dictionaries wird jedem Land ein Kontinent zugeordnet.
Diese Zuordnung ermöglicht es, im Rahmen der Hypothese zur Korrelation
zwischen wirtschaftlichem Wohlstand und Flüchtlingsaufnahme,
Länder innerhalb desselben Kontinents gezielt zu vergleichen
und regionale Unterschiede systematisch zu analysieren.
